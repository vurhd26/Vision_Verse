For 1st image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('1.png')

pixel_data = img.reshape((-1, 3))
pixel_data = np.float32(pixel_data)

plt.figure(figsize = (12,15))

plt.subplot(2,2,1)
plt.axis('off')
plt.title('Original image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
k = np.array([5, 10, 15])

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 500, 0.01)
j = 2
for i in k:
  compactness, labels, centers = cv2.kmeans(pixel_data, i, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
  centers = np.uint8(centers)
  segmented_data = centers[labels.flatten()]
  segmented_image = segmented_data.reshape((img.shape))
  plt.subplot(2, 2, j)
  plt.axis('off')
  plt.title(f'Segmented Image for k={i}')
  plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
  j = j + 1
```

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('boat.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thresh1 = cv2.threshold(img_gray, 50, 255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize = (12,15))

plt.subplot(2,2,1)
plt.axis('off')
plt.title('Original image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(2,2,2)
plt.axis('off')
plt.title('Otsu Thresholding')
plt.imshow(cv2.cvtColor(thresh1, cv2.COLOR_BGR2RGB))

ret, thresh2 = cv2.threshold(img_gray, 50, 255, cv2.THRESH_BINARY)
plt.subplot(2,2,3)
plt.axis('off')
plt.title('Normal Thresholding')
plt.imshow(cv2.cvtColor(thresh2, cv2.COLOR_BGR2RGB))
```
