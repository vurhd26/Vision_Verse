```python
import cv2
import matplotlib.pyplot as plt


img = cv2.imread('original.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)
plt.figure(figsize = (12,15))

#Original Image
plt.subplot(3,2,1)
plt.axis('off')
plt.title('Original image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

#Gaussain Filter
plt.subplot(3,2,2)
plt.axis('off')
plt.title('Gaussian filter')
plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_BGR2RGB))

img_median = cv2.medianBlur(img, 5)

#Median Filter
plt.subplot(3,2,3)
plt.axis('off')
plt.title('Median filter')
plt.imshow(cv2.cvtColor(img_median, cv2.COLOR_BGR2RGB))

img_bilateral = cv2.bilateralFilter(img, 15, 75, 75)

#Bilateral Filter
plt.subplot(3,2,4)
plt.axis('off')
plt.title('Bilateral filter')
plt.imshow(cv2.cvtColor(img_bilateral, cv2.COLOR_BGR2RGB))

Laplacian_filter = cv2.Laplacian(img_blur, -1, ksize=3,scale=1,delta=5)

#Laplacian Filter
plt.subplot(3,2,5)
plt.axis('off')
plt.title('Laplacian filter')
plt.imshow(cv2.cvtColor(Laplacian_filter, cv2.COLOR_BGR2RGB))

Box_filter = cv2.boxFilter(img, -1, (7, 7))

#Box Filter
plt.subplot(3,2,6)
plt.axis('off')
plt.title('Box filter')
plt.imshow(cv2.cvtColor(Box_filter, cv2.COLOR_BGR2RGB))
```

Here bilateral and gaussian filter are better than the other filters as the details are much more visible in both as compared to others. In gaussian and bilateral, bilateral is better as it smooths the image keeping in mind the edges.
