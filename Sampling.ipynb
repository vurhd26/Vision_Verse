```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def nearest_neighbor_downsampling(image, scale_factor):
    height, width, channels = image.shape
    new_height = int(height / scale_factor)
    new_width = int(width / scale_factor)
    downsampled_image = np.zeros((new_height, new_width, channels), dtype=np.uint8)
    for i in range(new_height):
        for j in range(new_width):
            original_i = i * scale_factor
            original_j = j * scale_factor
            downsampled_image[i, j] = image[original_i, original_j]
    return downsampled_image

image_path = 'PK Kelker Library.jpg'
original_image = cv2.imread(image_path)
downsampled_image_by_2 = nearest_neighbor_downsampling(original_image, 2)
downsampled_image_by_4 = nearest_neighbor_downsampling(original_image, 4)
downsampled_image_by_16 = nearest_neighbor_downsampling(original_image, 16)
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.title('Downsampled Image (Factor = 2)')
plt.imshow(cv2.cvtColor(downsampled_image_by_2, cv2.COLOR_BGR2RGB))
plt.subplot(2,2,3)
plt.title('Downsampled Image (Factor = 4)')
plt.imshow(cv2.cvtColor(downsampled_image_by_4, cv2.COLOR_BGR2RGB))
plt.subplot(2,2,4)
plt.title('Downsampled Image (Factor = 16)')
plt.imshow(cv2.cvtColor(downsampled_image_by_16, cv2.COLOR_BGR2RGB))
plt.show()
```
