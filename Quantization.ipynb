```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def uniform_quantization(image, n_levels):
    normalized_image = ((image / 255.0) * (n_levels - 1)).astype(np.uint8)
    quantized_image = np.round((normalized_image / (n_levels - 1)) * 255).astype(np.uint8)
    return quantized_image
image_path = 'PK Kelker Library.jpg'
quantization_levels = [2,4,8,16,32] #where each element represents the levels
original_image = cv2.imread(image_path)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.subplot(3, 2, 1)
plt.title('Original Image')
plt.imshow(original_image_rgb)
for i, levels in enumerate(quantization_levels):
        quantized_image = uniform_quantization(original_image, levels)
        plt.subplot(3, 2, i + 2)
        plt.title(f'Quantized Image ({levels} Colors)')
        plt.imshow(quantized_image)
        plt.show()
```
